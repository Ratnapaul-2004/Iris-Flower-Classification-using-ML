```python
import pandas as pd
```

```python
# Dataset 
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.feature_names
```

```python
X = pd.DataFrame(iris.data, columns = iris.feature_names) # sepal length, sepal width, petal length, petal width
y = pd.DataFrame(iris.target, columns = ['Class']) # flower class
```

```python
X
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn = KNeighborsClassifier(3)
Knn.fit(X,y)
```

```python
test_row = X.iloc[[87]]
y_pred = Knn.predict(test_row)
```

```python
print(y_pred)
```

```python
# Splitting dataset into train and test sets
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)
```

```python
Knn1 = KNeighborsClassifier(3)
Knn1.fit(X_train, y_train)
```

```python
y_predicted = Knn1.predict(X_test)
```

```python
print(y_predicted)
```

```python
# Model Performance Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_predicted, y_test))
```

```python
print(confusion_matrix(y_predicted, y_test))
```
